## 1. The Project Problem

On construction sites, it is difficult to monitor changes in the pressure, temperature, and humidity inside formwork in real time during concrete pouring. A sensor-based digital twin can quickly detect hazardous conditions and help workers make safer decisions.

In this exercise, we use sample sensor data and a 3D visualization app to answer the following questions.

- Where are the at-risk sensors in the current structure?
- Under what conditions do pressure or temperature warnings occur?
- What actions should be recommended to the site manager?
- What UI or algorithm improvements can each team make?
